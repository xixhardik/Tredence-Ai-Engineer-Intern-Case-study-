# Self-Pruning Neural Network Analysis
This notebook walks through the visual experiments of our self-pruning neural network architecture.
We explore the tradeoff between standard classification loss and the L1 Sparsity Penalty on the gating mechanism.

### The Mathematical Formulation
Our loss function is defined as:
$$ \mathcal{L}_{total} = \mathcal{L}_{CE}(y, \hat{y}) + \lambda \sum_{i} |\sigma(g_i)| $$

Where:
- $\mathcal{L}_{CE}$ is the standard Cross-Entropy Loss.
- $g_i$ represents the learnable `gate_scores`.
- $\sigma$ is the sigmoid function.
- $\lambda$ dictates the sparsity vs accuracy tradeoff.

### Key Findings
1. **The 'Corner Solution' Effect**: The L1 penalty forces the optimizer to push smaller gates exactly to 0, effectively severing those connections.
2. **Lambda Tradeoff**: As $\lambda$ increases, the model happily sacrifices classification accuracy to drastically drop its parameter count.
3. **Dynamic Architecture**: The network discovers its optimal size *during* training, rather than requiring expensive post-training pruning loops!
